```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load data
raw = fetch_california_housing()
data = pd.DataFrame(raw.data, columns=raw.feature_names)

# Tambahkan kolom target
data['median_house_value'] = raw.target

# Baru deh ini jalan:
X = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
```

```python
train_data['total_rooms'] = np.log(data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(data['total_bedrooms'] + 1)
train_data['population'] = np.log(data['population'] + 1)
train_data['households'] = np.log(data['households'] + 1)
```

```python
if 'ocean_proximity' in train_data.columns:
    ocean_dummies = pd.get_dummies(train_data['ocean_proximity'])
    train_data = train_data.join(ocean_dummies)
    train_data = train_data.drop(['ocean_proximity'], axis=1)
else:
    print("Kolom 'ocean_proximity' tidak ditemukan.")
```
